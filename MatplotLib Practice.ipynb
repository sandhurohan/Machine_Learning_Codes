```python
company=['GOOGLE','AMAZON','ZOMATO','SWIGGY']
revenue=[154,584,652,784]
profit=[70,54,98,12]
```

```python
plt.xlabel('Companies')
plt.ylabel('Revenue')
plt.title('Stocks List')
plt.bar(company,revenue,label='Revenue',width=0.4,edgecolor='yellow')
plt.bar(company,profit,label='Profit',width=0.4,edgecolor='re')
plt.legend()
plt.show()
```
